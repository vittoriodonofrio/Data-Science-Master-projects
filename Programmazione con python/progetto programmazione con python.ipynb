```python
def add_products ():
    
    '''This function is used to add a product to the warehouse'''
    
    
    with open("warehouse.csv", "r") as warehouse:
        warehouse.seek(0)
        csv_reader=csv.reader(warehouse)
        for row in csv_reader:
            prod[f"{row[0]}"] = [int(row[1]) , float(row[2]), float(row[3]) ]
        
    
    product_name = input("Nome del prodotto: ")
    quantity=""
    purchase_price=""
    selling_price=""
    
    
    if "**" in product_name:
        quantity=int(product_name.split("**")[2])
        product_name=product_name.split("**")[0]
    else:  
        while type(quantity)!=int or quantity <= 0:
            
            try:
                quantity = int(input("Quantità: "))
                if quantity<=0:
                    raise ValueError ("")
            except ValueError:
                print("Inserisci un numero intero positivo come quantità!")
                
                
    if product_name in set(prod):
        prod[product_name] [0] += quantity
    else:
        while type(purchase_price)!=float or purchase_price <= 0:
            
            try:
                purchase_price = float(input("Prezzo di acquisto: "))
                if purchase_price <= 0:
                    raise ValueError ("")
            except ValueError:
                print("Inserisci un numero positivo!")
                
        while type(selling_price)!=float or selling_price <=0:
            
            try:
                selling_price = float(input("Prezzo di vendita: "))
                if selling_price <= 0:
                    raise ValueError ("")
            except ValueError:
                print("Inserisci un numero positivo!")
            
            
        prod[product_name] = [quantity, purchase_price, selling_price]
        
    with open("warehouse.csv", "w+", newline='') as warehouse:
        warehouse.seek(0)
        csv_writer = csv.writer(warehouse)
        for row in list(prod):
            csv_writer.writerow([row, prod[f"{row}"][0], prod[f"{row}"][1], prod[f"{row}"][2] ])
            
            
    print(f"Aggiunto {quantity}x {product_name}")
```

```python
class sales:
    
    '''This class is used to record a sale and update profits'''
    
    def __init__ (self, product_name, quantity, prod, profits, total, note):
        self.product_name=product_name
        self.quantity=quantity
        self.prod=prod
        self.profits=profits
        self.total=total
        self.note=note
    
    
    def sold_products(self):
        
        '''Record a sale'''
        
        if self.product_name in set(self.prod):
            if(self.quantity<=self.prod[self.product_name][0]):
                self.prod[self.product_name][0] -= self.quantity
                self.note = self.note + f"\n{self.quantity}x{self.product_name}: {prod[product_name][2]}"
                self.total += self.quantity*self.prod[product_name][2]
            else:
                print("Non ce ne sono abbastanza in magazzino")
        else:
            print("Prodotto non disponibile in magazzino")
        return self.note, self.total
            
    def profits_update(self):
        
        '''Update profits'''
        
        self.profits["gross profit"][0] += self.quantity*self.prod[self.product_name][2]
        self.profits["net profit"][0] +=self.quantity*self.prod[self.product_name][2]-self.quantity*self.prod[self.product_name][1]
        with open("profits.csv", "w+", newline='') as profit:
                profit.seek(0)
                csv_writer = csv.writer(profit)
                csv_writer.writerow([ round(float( profits["gross profit"][0] ) ,2), round(float( profits["net profit"][0]),2)  ])
```

```python
def help_ ():
    print(f"I comandi disponibili sono i seguenti:  \n aggiungi: aggiunge un prodotto al magazzino \n elenca: elenca i prodotti in magazzino \n vendita: registra una vendita effettuata \n profitti: mostra i profitti totali \n aiuto: mostra i possibili comandi \n chiudi: esci dal programma"   
         )
```
